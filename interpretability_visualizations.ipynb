```python
from constants import *

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
import os
```

```python
RESULTS_FOLDER = "results"
INTERPRETABILITY_SOURCE_FOLDER = os.path.join(RESULTS_FOLDER, "feature_ablation")
VISUALIZATIONS_FOLDER = os.path.join(RESULTS_FOLDER, "visualizations")
MODEL = "granitecode3b"
TASK = "int_cmp"
```

```python
lang1 = sorted(LANGS.keys())[0]
input_tokens = None
lang_results = {v: None for k, v in LANGS.items() if k != lang1}
for lang2 in LANGS.keys():
    if lang1 == lang2:
        continue
    filename = f"{MODEL}-{TASK}-{lang1.lower()}-{lang2.lower()}.json"
    results = json.load(
        open(os.path.join(INTERPRETABILITY_SOURCE_FOLDER, filename), "r")
    )
    if input_tokens is None:
        input_tokens = [
            s.encode("ascii", errors="ignore").decode() for s in results["input_tokens"]
        ]
    lang_results[LANGS[lang2]] = results["seq_attr"]
collected_results = pd.DataFrame(lang_results)
collected_results.index = input_tokens
# collected_results = collected_results.reset_index(names=f"{LANGS[lang1]} code")
print(
    f"Visualizing code conversion interpretability for model `{MODELS[MODEL]}` on the task `{TASK}` from language `{LANGS[lang1]}`"
)
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
_ = sns.heatmap(
    collected_results,
    center=0,
    robust=True,
    cmap="RdYlGn",
    annot=True,
    fmt=".2f",
    linewidths=1,
    ax=ax,
)
ax.xaxis.set_ticks_position("top")
_ = ax.set_ylabel(f"{LANGS[lang1]} code for `{TASK}` task")
_ = ax.set_title(f"Code Conversion interpretability for {MODELS[MODEL]}")
fig.savefig(
    os.path.join(VISUALIZATIONS_FOLDER, f"{MODEL}-{TASK}-{lang1.lower()}.png"),
    bbox_inches="tight",
)
```
