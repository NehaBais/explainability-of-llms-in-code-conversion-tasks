```python
from matplotlib.font_manager import FontProperties
from constants import *

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
import os
```

```python
RESULTS_FOLDER = "results"
INTERPRETABILITY = "shapley_sampling"
INTERPRETABILITY_SOURCE_FOLDER = os.path.join(RESULTS_FOLDER, INTERPRETABILITY)
VISUALIZATIONS_FOLDER = os.path.join(RESULTS_FOLDER, "visualizations")
MODEL = "yicoder9b"
TASK = "logic_ops"
MONOSPACE = FontProperties(family='monospace')
```

```python
def snake_to_capitalize(text):
    text = text.split("_")
    return " ".join([word.capitalize() for word in text])
```

```python
def crop_text(text, max_len):
    crop_len = 50
    if max_len <= crop_len:
        return text.ljust(max_len)
    else:
        if len(text) <= crop_len:
            return text.ljust(crop_len)
        return f"{text[:crop_len - 4]} ..."
```

```python
lang1 = sorted(LANGS.keys())[0]
input_tokens = None
lang_results = {v: None for k, v in LANGS.items() if k != lang1}
for lang2 in LANGS.keys():
    if lang1 == lang2:
        continue
    filename = f"{MODEL}-{TASK}-{lang1.lower()}-{lang2.lower()}.json"
    filename = os.path.join(INTERPRETABILITY_SOURCE_FOLDER, filename)
    if not os.path.exists(filename):
        continue
    results = json.load(open(filename, "r"))
    if input_tokens is None:
        max_len = max([len(s) for s in results["input_tokens"]])
        input_tokens = [
            crop_text(s.encode("ascii", errors="ignore").decode(), max_len)
            for s in results["input_tokens"]
        ]
    lang_results[LANGS[lang2]] = results["seq_attr"]
lang_results = {k: v for k, v in lang_results.items() if v is not None}
collected_results = pd.DataFrame(lang_results)
collected_results.index = input_tokens
print(
    f"Visualizing code conversion interpretability for model `{MODELS[MODEL]}` on the task `{TASK}` from language `{LANGS[lang1]}`"
)
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
_ = sns.heatmap(
    collected_results,
    center=0,
    robust=True,
    cmap="RdYlGn",
    annot=True,
    fmt=".2f",
    linewidths=1,
    ax=ax,
)
ax.xaxis.set_ticks_position("top")
_ = ax.set_ylabel(f"{LANGS[lang1]} code for `{TASK}` task")
_ = ax.set_yticklabels(ax.get_yticklabels(), fontproperties=MONOSPACE)
fig.suptitle(
    f"{snake_to_capitalize(INTERPRETABILITY)} interpretability for {MODELS[MODEL]}"
)
fig.savefig(
    os.path.join(
        VISUALIZATIONS_FOLDER, f"{INTERPRETABILITY}-{MODEL}-{TASK}-{lang1.lower()}.png"
    ),
    bbox_inches="tight",
)
```
